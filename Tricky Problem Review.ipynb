### Problem Statement

Given a linked list with integer data, arrange the elements in such a manner that all nodes with even numbers are placed after odd numbers. **Do not create any new nodes and avoid using any other data structure. The relative order of even and odd elements must not change.** 

**Example:**
* `linked list = 1 2 3 4 5 6`
* `output = 1 3 5 2 4 6`

````python
# Example 1 2 3 4 5 -> 1 3(prev2) 2 4(runner1) 5(runner2) 
# 2 1 -> 1 2 4 2 6 1
def even_after_odd(head):
    """
    :param - head - head of linked list
    return - updated list with all even elements are odd elements
    """
    if head is None:
        return head
    prev = head 
    runner1 = head
    runner2 = head.next
    while(runner1 and runner2):
        if runner1.data % 2 == 0 and runner2.data % 2 != 0:
            # Two cases
            if prev.data % 2 == 0:
                runner1.next = runner1.next.next
                runner2.next = prev
                head = runner2
                prev = runner2
                runner2 = runner1.next
            else:
                runner1.next = runner1.next.next
                temp = prev.next
                prev.next = runner2
                runner2.next = temp
                prev = runner2
                runner2 = runner1.next
        else:
            runner1 = runner1.next
            runner2 = runner2.next
    return head
            
                
        
            
````

## Problem Statement

Given a linked list, swap the two nodes present at position `i` and `j`. The positions are based on 0-based indexing.

**Note:** You have to swap the nodes and not just the values. 

**Example:**
* `linked_list = 3 4 5 2 6 1 9`
* `positions = 3 4`
* `output = 3 4 5 6 2 1 9`

**Explanation:** 
* The node at position 3 has the value `2`
* The node at position 4 has the value `6`
* Swapping these nodes will result in a final order of nodes of `3 4 5 6 2 1 9`

````python
# Solution

def swap_nodes(head, left_index, right_index):
    # if both the indices are same
    if left_index == right_index:
        return head
    
    
    left_previous = None
    left_current = None

    right_previous = None
    right_current = None

    count = 0
    temp = head
    new_head = None

    # find out previous and current node at both the indices
    while temp is not None:
        if count == left_index:
            left_current = temp
        elif count == right_index:
            right_current = temp
            break

        if left_current is None:
            left_previous = temp
        right_previous = temp
        temp = temp.next
        count += 1
    # 1(leftcurr) 2 3 4(rightcurr)
    # 3 - 1 - 2 - 3 temp: 2 - 3-1- None
    right_previous.next = left_current # 1-2-1(left curr)-2-3-4
    # MIGHT CREATE A LOOP
    temp = left_current.next # 2-1-2 2 3 4
    # BREAK THE LOOP
    left_current.next = right_current.next # 2 - 1 - 4  1 - none 

    # if both the indices are next to each other
    print(left_index != right_index)
    if left_index != right_index:
        right_current.next = temp # 3 - 2 - 1 - 4
        
    # if the node at first index is head of the original linked list
    if left_previous is None:
        new_head = right_current
    else:
        left_previous.next = right_current
        new_head = head

    return new_head
````

````python
class Solution:
    def rotate(self, matrix) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        # O(N^2)
        for i in range(len(matrix)//2):
            for j in range(i,len(matrix)-1-i):
                top = matrix[i][j]
                matrix[i][j] = matrix[len(matrix)-1-j][i] # top -> left
                matrix[len(matrix)-1-j][i] = matrix[len(matrix)-1-i][len(matrix)-1-j] #left -> bottom
                matrix[len(matrix)-1-i][len(matrix)-1-j] = matrix[j][len(matrix)-1-i] #bottom -> right
                matrix[j][len(matrix)-1-i] = top #right -> top
````
